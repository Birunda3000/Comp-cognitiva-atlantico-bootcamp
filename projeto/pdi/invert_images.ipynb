```python
import cv2
from pdi_utils import show_image
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
import os
import pathlib
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
```

```python
DATADIR = '../../../_DATA_/Folhas_segmented'
OUTDIR = '../../../_DATA_/Folhas_invert'
DATADIR = pathlib.Path(DATADIR)
files = os.listdir(DATADIR)
files.sort(reverse=True)
```

```python
for img in tqdm(files):
    path = os.path.join(DATADIR,img)
    out_path = os.path.join(OUTDIR,img+'-resize-semi-seg-inverted.jpg')
    imagem = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    imagem = cv2.bitwise_not(imagem)
    #plt.imshow(np.array(imagem/255.0), cmap='gray')
    show_image(np.array(imagem/255.0))
    #plt.show()
    #show_image(binary)
    #plt.imshow(binary, cmap='gray')
    #plt.show()
    cv2.imwrite(out_path, imagem)
    print('-------------------')
```
