```python
import cv2
from pdi_utils import show_image
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
```

```python
imagem = cv2.imread('download2.jpg')
imagem = cv2.resize(imagem, (1920,1080))
imagem = rgb2gray(imagem)
show_image(imagem, 'Binary image')
```

```python
thresh = 0.8
binary = imagem > thresh
show_image(binary, f'Binary image - threshold - {thresh} + manual coloring', save=True)
#cv2.imwrite("saida.jpg", binary)
```

```python
for i in range(0,100,5):
    thresh = i/100
    binary = imagem > thresh
    show_image(binary, f'Binary image - threshold - {thresh} + manual coloring')
```

```python
from skimage.filters import try_all_threshold
import matplotlib.pyplot as plt
fig, ax = try_all_threshold(imagem, verbose=False)

# Show the resulting plots
plt.show()
```
