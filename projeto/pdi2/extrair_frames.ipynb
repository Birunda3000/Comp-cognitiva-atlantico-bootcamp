```python
import cv2
import os
import pathlib

def FrameCapture(path_in, path_out, n_video, PULAR_FRAMES=1): 
    vidObj = cv2.VideoCapture(path_in) 
    count = 1
    success = 1
    global count_global
    while success: 
        success, image = vidObj.read()
        if success:
            if count % PULAR_FRAMES == 0:
                out = path_out + '/video-' + str(n_video) + '-frame-' + str(count) + '-img-' + str(count_global) + '.jpg'
                out = pathlib.Path(out)
                out = str(out)
                cv2.imwrite(out, image)
                print("SUCESSO - " + out)
                count_global += 1
            count += 1
    
    print("Frames extracted")

count_global = 0

DATADIR = "videos"
DATADIR = pathlib.Path(DATADIR)

OUTDIR = "frames"
OUTDIR = pathlib.Path(OUTDIR)

for folder in os.listdir(DATADIR):
    
    class_path = os.path.join(DATADIR,folder)
    num_video = 0
    for video in os.listdir(class_path):
        num_video += 1
        #FrameCapture(os.path.join(class_path,video), os.path.join(OUTDIR,folder),num_video)

        if folder == 'Diego':
            print('6')
            FrameCapture(os.path.join(class_path,video), os.path.join(OUTDIR,folder),num_video, PULAR_FRAMES=6)

        elif folder == 'Erick' or folder == 'Jeronimo':
            print('2')
            FrameCapture(os.path.join(class_path,video), os.path.join(OUTDIR,folder),num_video, PULAR_FRAMES=2)

        elif folder == 'Pedro' or folder == 'Walter':
            print('1')
            FrameCapture(os.path.join(class_path,video), os.path.join(OUTDIR,folder),num_video, PULAR_FRAMES=1)
        else:
            print('-----------------------------ERRO-------------------------------------')



    count_global = 0
```
