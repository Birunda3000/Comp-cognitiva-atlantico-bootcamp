```python
def canny_p(img):
    
    can = cv2.Canny(img, 60, 100)

    img_rgb = cv2.cvtColor(can, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb, cmap='gray')
    plt.show()
    return can
```
